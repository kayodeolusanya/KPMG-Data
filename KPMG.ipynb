```python
# Standardizing the gender column
gender_value_mapping = {
    'F': 'Female',
    'Male': 'Male',
    'Female': 'Female',
    'Femal': 'Female',
    'M': 'Male',
    'U': 'Non-Binary'

}
```

```python
df['NewCustomerList']['gender'] =df['NewCustomerList']['gender'].replace(gender_value_mapping)
df['NewCustomerList']['gender'].unique()
```

```python
df['CustomerDemographic']['gender'] =df['CustomerDemographic']['gender'].replace(gender_value_mapping)
df['CustomerDemographic']['gender'].unique()
```

```python
state_value_mapping = {
    'Victoria': 'VIC',
    'New South Wales': 'NSW',
    'VIC': 'VIC',
    'QLD': 'QLD',
    'NSW': 'NSW'

}
```

```python
df['NewCustomerList']['state'] =df['NewCustomerList']['state'].replace(state_value_mapping)
df['NewCustomerList']['state'].unique()
```

```python
df['CustomerAddress']['state'] =df['CustomerAddress']['state'].replace(state_value_mapping)
df['CustomerAddress']['state'].unique()
```

```python
for sheet_name in df:
  numerical_values = df[sheet_name].select_dtypes(include=['float64','int64']).columns
  categorical_values = df[sheet_name].select_dtypes(include=['object']).columns
  df[sheet_name][numerical_values] =   df[sheet_name][numerical_values].fillna(df[sheet_name][numerical_values].median())
  df[sheet_name][categorical_values] =   df[sheet_name][categorical_values].fillna(df[sheet_name][categorical_values].mode())
```
